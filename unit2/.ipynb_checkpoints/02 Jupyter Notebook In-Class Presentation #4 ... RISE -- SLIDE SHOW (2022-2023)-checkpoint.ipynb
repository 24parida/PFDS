
# CREATING A SLIDE SHOW

<font size="4">In the notebook menu, the “View” option contains a “Cell Toolbar” sub-menu that gives you access to the metadata for each cell. If you select the “Slideshow” preset, you will see in the right corner of each cell a little box where you can select the cell type.

You can choose between the following types:</font>

<font size="4">**slide:** this cell is the beginning of a new slide
<p>&nbsp;</p>
    
**subslide:** this cell is the beginning of a new subslide; that is to say, a new slide, but that reveal.js will display below the previous one instead of on the right<p>&nbsp;</p>**fragment:** this is to split the contents of one slide into pieces; a cell marked as a fragment will create a break inside the slide; it will not show up right away, you will need to press Space one more time to see it. 
***
**skip:** this cell is ignored altogether in reveal mode, it will not appear either in the main view, nor in the speaker view.
notes: similarly, this cell is marked to be discarded from the main view, but is meant to appear in the speaker view.

    
**notes:** similarly, this cell is marked to be discarded from the main view, but is meant to appear in the speaker view.
</font>

# RUNNING A SLIDE SHOW

<font size="4">Once enabled, the RISE Jupyter extension displays a new button (“Enter/Exit Live Reveal Slideshow”) in the toolbar, (also activable with Alt-r by default).

This starts the slideshow; you can return to normal notebook edition by either pressing Alt-r again, or by clicking on the cross-shaped icon on the upper right corner of each slide.,/font>

# NAVIGATION

<font size="4">It is highly recommended to use mainly SpaceBar to go forward, and Shift-SpaceBar to go backward (or the visual controller in the slideshow right bottom corner). This will follow the course of the presentation no matter what the detailed structure is (slides, subslides, fragments…).

In contrast, right and left arrows can have a confusing behaviours with respect to these 3 structural entities. Besides, up and down arrows are reserved to interact with notebook cells and cannot be used to navigate the slides, instead you can use pgup and pgdown.</font>

# SELECTION & EVALUATION

<font size="4">Essentially, when a code cell appears in the presentation, you simply need to press Shift-Enter to run it. This will move to the next cell if it is already displayed.

The default behaviour for RISE is to select the first code cell when a new slide or fragment shows up. This way, if your presentation has only markdown cells, you will not be bothered with cells being selected; on the other hand when yo do have code cells, you can run the entire slideshow by just using Space and Shift-Enter as appropriate.</font>

### Title

### Menu bar

<font size="4">This is similar to any other desktop application; it allows you to do different actions with the Notebook.</font>

### Toolbar

<font size="4">This is located below the menu bar and contains little icons to perform some commonly used actions ... e.g. saving a file, cutting cells, pasting cells, moving cells, etc.</font>

### Mode indicator

<font size="4">This is located on the right hand side of the menu bar.  The Notebook has two modes -- Edit mode and Command mode. This will be discussed in greater detail below.</font>

### Kernal indicator

<font size="4">This display's the status of the system's computation progress.  To interrupt the computations that are in progress, you can use the stop button located on the toolbar.</font>

### Notification area

<font size="4">This area will display messages, suchas **saving the file, or **interrupting the kernal**, etc. -- and you will see the actions that you are performing in the notification area.</font>

<font size="4">These Jupyter Notebook user interfaces can be very useful as they display important information and notifications while you work on the Notebook.

If we go to the **Help** tab and click on the first option, **User Interface Tour**, we can get a quick look at what is involved with these important interfaces of Jupyter Notebook.</font>

## C. Keyboard Input Modes 
<p>&nbsp;</p>
<font size="4">As mentioned above, the Notebook has two modes -- Edit mode and Command mode.
<p>&nbsp;</p>
Command mode has many shortcut keys that you can use.  In this mode, no icon is displayed in the Mode indicator are, and the actions that you can perform in this mode are actions that have to do with the file itself suchas saving the file, copying and pasting cells, etc.
<p>&nbsp;</p>
Edit mode allows you to write code or text in a cell, if you are in a cell.  When you are in Edit mode, you will see a little pencil in the indicator area.
<p>&nbsp;</p>
The border of a selected cell will be GREEN when you are in Edit mode.
<p>&nbsp;</p>
The border of a selected cell will be BLUE when you are in Command mode.
<p>&nbsp;</p>
To go back to Command mode from Edit mode, you can press the Esc key.
<p>&nbsp;</p>
To go back to Edit mode from Command mode, you can press the Enter key.
<p>&nbsp;</p>    
There are many helpful keyboard shortcuts that you can use and there is a list in the **Help** menu.
</font>
