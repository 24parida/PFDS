<div class="alert alert-block alert-success"><b>Question #10:</b>: 

Step #1: Write code that asks the user for their dog's name and then asks the user for their dog's age in years.
<p>&nbsp;</p>    
Step #2: Write a formatted print statement to print the dog's name and age in dog years (to the nearest tenth).
<p>&nbsp;</p>    
You will need to use the formula below to calculate the age of the dog in dog years.
<p>&nbsp;</p   
<center>dog_years = (10.5 * 2) + (4 * (age - 2))</center>
<p>&nbsp;</p>    
Run your code for a dog named Charlie that is 10.5 years old.  If you have written the proper code, you shoulget the following output:
    


</div>


```python
dog_name = input("What is your dog's name? ")
capital_first_name = dog_name.capitalize()
dog_age = float(input("How old (in years) is your dog? "))
dog_years = (10.5 * 2) + (4 * (dog_age - 2))

print(f"{capital_first_name} is {dog_years} old in dog years.")
```
