```python
def find_volume(r,h):
    return math.pi * (r**2) * h
def find_surface_area(r, h):
    return (2 * math.pi * (r**2)) + (2 * math.pi * r * h)
r = int(input("Input the radius: "))
h = int(input("Input the height: "))
print(f"Volume: {find_volume(r,h):.2f}")
print(f"Surface Area: {find_surface_area(r,h):.2f}")
```
