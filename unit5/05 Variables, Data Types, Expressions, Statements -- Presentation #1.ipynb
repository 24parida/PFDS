<b><font size="4">II. Type Conversion Functions</font></b></div>

Sometimes it is necessary to convert values from one type to another. Python provides a few simple functions that will allow us to do that. The functions <code style="background:lightyellow;color:black">int</code>, <code style="background:lightyellow;color:black">float</code> and <code style="background:lightyellow;color:black">str</code> will (attempt to) convert their arguments into types int, float and str respectively. 

We call these **type conversion** functions.

The <code style="background:lightyellow;color:black">int</code> function can take a floating point number or a string, and turn it into an int. For floating point numbers, it discards the decimal portion of the number - a process we call **truncation towards zero on the number line**. 

```python
# Examples of int type conversion

print(2.7183, type(2.7183))
print(int(2.7183), type(int(2.7183)))      # This doesn't round to the closest int! It truncates toward 0 on the number line!

print('\n', 5.9999, type(5.9999))
print(int(5.9999), type(int(5.9999)))      # This doesn't round to the closest int! It truncates toward 0 on the number line!
         
print('\n', 2.0, type(2.0))
print(int(2.0), type(int(2.0)))
      
print('\n', -2.7183, type(-2.7183))
print(int(-2.7183), type(int(-2.7183)))     # Note that the result is closer to zero

print('\n', "2021", type("2021"))
print(int("2021"), type(int("2021")))       # parse a string to produce an int

print('\n', 75, type(75))
print(int(75), type(int(75)))               # int even works on integers
```

```python
# Invalid literal for int conversion

print('23 bottles', type('23 bottles'))
print(int('23 bottles'))
```

- The last case shows that a string has to be a syntactically legal number, otherwise you’ll get one of those pesky runtime errors.

- The type converter <code style="background:lightyellow;color:black">str</code> turns its argument into a string. 

- Remember that when we print a string, the quotes are removed. However, if we print the type, we can see that it is definitely *str*.

```python
# Examples of str type conversion

print(53)
print(type(53))

print('\n', str(53))
print(type(str(53)))

print('\n', 204.71)
print(type(204.71))

print('\n', str(204.71))
print(type(str(204.71)))
```

<b><font size="4">III. VARIABLES</font></b></div>

One of the most powerful features of a programming language is the ability to manipulate **variables**. A variable is a name that refers to a value.

**Assignment statements** create new variables and also give them values to refer to.

```python
# This is a code cell ... what happens when you run it?

message = "I hope that you are having a great day!"
n = 27
e = 2.7183
```

This example makes three assignments. 

The first assigns the string value <code style="background:lightyellow;color:black">"I hope that you are having a great day!"</code> to a new variable named <code style="background:lightyellow;color:black">message</code>. 

The second gives the integer <code style="background:lightyellow;color:black">27</code> to <code style="background:lightyellow;color:black">n</code>.

The third assigns the floating-point number <code style="background:lightyellow;color:black">2.7183</code> to a variable called <code style="background:lightyellow;color:black">e</code>.

The assignment token, <code style="background:lightyellow;color:black">=</code>, should not be confused with equality (we will see later that equality uses the <code style="background:lightyellow;color:black">==</code> token). 

The assignment statement links a name, on the left hand side of the operator, with a value, on the right hand side. 

This is why you will get an error if you enter the following:

```python
# Run the code in this cell

27 = n
```

```python
# Use this cell for the tip below

n = 17
```

<div class="alert alert-block alert-warning"><b>Tip</b>

When reading or writing code, say to yourself ...

“n is assigned 17” or “n gets the value 17” or “n is a reference to the object 17” or “n refers to the object 17”.

Don’t say “n equals 17”.</div>

A common way to represent variables on paper is to write the name with an arrow pointing to the variable’s value. 

This kind of figure, known as a **reference diagram**, is often called a **state snapshot** because it shows what state each of the variables is in at a particular instant in time. (Think of it as the variable’s state of mind). 

This diagram shows the result of executing the assignment statements shown above.

If you ask Python to evaluate a variable, it will produce the value that is currently linked to the variable. 

In other words, evaluating a variable will give you the value that is referred to by the variable.

```python
# Run the code in this cell

message = "I hope that you are having a great day!"
n = 27
e = 2.7183

print(message)
print(n)
print(e)
```

In each case the result is the value of the variable.

Variables also have types; again, we can ask the interpreter what they are.

```python
# Run the code in this cell

message = "I hope that you are having a great day!"
n = 27
e = 2.7183

print(type(message))
print(type(n))
print(type(e))
```

The type of a variable is the type of the object it currently refers to.

We use variables in a program to “remember” things, like the current score at the football game. But variables are variable. This means they can change over time, just like the scoreboard at a football game. You can assign a value to a variable, and later assign a different value to the same variable.
